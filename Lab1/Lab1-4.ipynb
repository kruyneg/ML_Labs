2. Предложите способ заменить отсутствующие значения из пункта 1 и реализуйте его. Способ нужно аргументировать(сказать почему это достаточно хорошее приближение)

````python
dataset["full_sq"].isnull().sum()
````

Информация о полной площади жилого помещения присутствует во всех записях. Найдем среднее соотношение `p` жилой площади и полной, и заменим пропуски на `full_sq * p`.

````python
normal_data = dataset[dataset["life_sq"] <= dataset["full_sq"]] # данные, в которых жилая площадь не превышает полную
p = normal_data["life_sq"] / normal_data["full_sq"]
p = p.mean()
print(p)

dataset["life_sq"] = dataset["life_sq"].fillna(dataset["full_sq"] * p)
dataset["life_sq"].isnull().sum()
````

7. Постройте график изменения медианной цены на квартиру менее 50 кв. метров за период с 2011 до 2014 года*

\* Задание на доп. балл

````python
dataset["date"] = pd.to_datetime(dataset["timestamp"])
dataset["date"] = dataset["date"].dt.to_period('M')

mask = (dataset["full_sq"] < 50) & (dataset["date"].dt.year >= 2011) & (dataset["date"].dt.year <= 2014)

median_prices = dataset[mask].groupby('date')['price_doc'].median()

plt.figure(figsize=(10, 5))
plt.xticks(rotation=70)

sns.lineplot(x=median_prices.index.astype(str), y=median_prices.values)
````
